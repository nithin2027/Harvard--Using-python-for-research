```python
from scipy.stats import bernoulli
import networkx as nx
import matplotlib.pyplot as plt
```

```python
bern_mat = []
for i in range(0,20):
    bern_mat.append((bernoulli.rvs(0.2)))
print(bern_mat)
```

```python
N = 20
p = 0.2
#create an empty graph
G = nx.Graph()
#add all nodes in the graph
G.add_nodes_from(range(N))
#loop over all pairs of nodes
for node1 in G.nodes():
    for node2 in G.nodes():
        if bernoulli.rvs(p=p):
            #add an edge with prob p
            G.add_edge(node1, node2)
```

```python
G.number_of_nodes()
```

```python
nx.draw(G, with_labels = 'True', node_color = 'orange', edge = 'grey')
plt.show()
```

```python
N = 20
p = 0.2
#create an empty graph
G = nx.Graph()
#add all nodes in the graph
G.add_nodes_from(range(N))
#loop over all pairs of nodes
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 < node2 and bernoulli.rvs(p=p):    #If not already connected
            #add an edge with prob p
            G.add_edge(node1, node2)

nx.draw(G, with_labels = 'True', node_color = 'magenta', edge = 'grey')
plt.show()
```

```python
def er_graph(N,p):
    '''
    Generate an ER graph
    '''
    G = nx.Graph()
    #add all nodes in the graph
    G.add_nodes_from(range(N))
    #loop over all pairs of nodes
    for node1 in G.nodes():
        for node2 in G.nodes():
            if node1 < node2 and bernoulli.rvs(p=p):
                
                #add an edge with prob p
                G.add_edge(node1, node2)
    return G
```

```python
nx.draw(er_graph(50, 0.08), node_size = 40, node_color = 'pink')
plt.show()
```

```python
nx.draw(er_graph(10 ,0), node_size = 40, node_color = 'black')
plt.show()
```

```python
nx.draw(er_graph(10 ,1), node_size = 40, node_color = 'violet')
plt.show()
```
