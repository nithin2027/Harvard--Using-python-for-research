```python
from scipy.stats import bernoulli
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
A1 = np.loadtxt('adj_allVillageRelationships_vilno_1.csv', delimiter = ',')
A2 = np.loadtxt('adj_allVillageRelationships_vilno_2.csv', delimiter = ',')
G1 = nx.to_networkx_graph(A1)
G2 = nx.to_networkx_graph(A2)
```

```python
nx.connected_component_subgraphs(G1)
```

```python
gen = nx.connected_component_subgraphs(G1)
g = gen.__next__()
type(g)
```

```python
g.number_of_nodes()
```

```python
G1.number_of_nodes()
```

```python
#LCC Largest connected component
G1_LCC = max(nx.connected_component_subgraphs(G1), key = len)
G2_LCC = max(nx.connected_component_subgraphs(G2), key = len)
```

```python
# compute the propotion of nodes that lie in 
# the largest connected components for the graphs

print('The propotion of the largest component for graph 1 is: ' + str( G1_LCC.number_of_nodes()/G1.number_of_nodes()))
print('The propotion of the largest component for graph 2 is: ' + str( G2_LCC.number_of_nodes()/G2.number_of_nodes()))
```

```python
plt.figure()
nx.draw(G1_LCC, node_color = "red", edge_color = "gray", node_size = 20)
plt.show()
```

```python
plt.figure()
nx.draw(G2_LCC, node_color = "green", edge_color = "gray", node_size = 20)
plt.show()
```
