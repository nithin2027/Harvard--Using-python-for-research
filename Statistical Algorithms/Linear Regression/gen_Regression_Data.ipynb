```python
import numpy as np 
import scipy.stats as ss
import matplotlib.pyplot as plt 
```

```python
n = 100
beta_0, beta_1 = 5,2
```

```python
np.random.seed(1)
x = 10* ss.uniform.rvs(size = n)    # Gen inputs
y = beta_0 + beta_1*x + ss.norm.rvs(loc = 0, scale = 1, size = n)   #Gen corresponding output:: f(x) = y, the ss.norm is for some noise
```

```python
plt.plot(x, y, 'o', ms = 5)
xx = np.array([0,10])
plt.plot(xx, (beta_0 + beta_1 * xx))
plt.xlabel('$x$')
plt.ylabel('$y$')
```
